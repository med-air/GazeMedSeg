```python
import pandas as pd
```

```python
df = pd.read_csv("kvasir_fixation_preprocessed_old.csv")
df
```

```python
df_gaze = df[
    [
        "images",
        "CURRENT_FIX_INDEX",
        "CURRENT_FIX_X",
        "CURRENT_FIX_Y",
        "CURRENT_FIX_DURATION",
        "CURRENT_FIX_PUPIL",
        "CURRENT_FIX_START",
    ]
]
```

```python
df_imgs = pd.read_csv("/media/yesindeed/DATADRIVE1/data/Kvasir-SEG/gaze/exp_resources/all_exp_list.csv")

for i, (name, group) in enumerate(df_gaze.groupby("images")):
    img_file = group["images"].tolist()[0]

    org_img_size = tuple(df_imgs.loc[df_imgs["images"] == img_file][["org_img_height", "org_img_width"]].values[0])

    df_gaze.loc[df_gaze["images"] == img_file, "IMAGE_HEIGHT"] = int(org_img_size[0])
    df_gaze.loc[df_gaze["images"] == img_file, "IMAGE_WIDTH"] = int(org_img_size[1])
```

```python
df_gaze = df_gaze.rename(columns={"images": "IMAGE"})

df_gaze
```

```python
df_gaze.to_csv("kvasir_fixation.csv", index=False)
```
